## 5.4. Deteksi Obyek Pada Video 
### 5.4.1. Deteksi Kendaraan Di Jalan Tol Versi Google Colab

````python
#Listing  5.6. 
import cv2 
import torch
import numpy as np 
from IPython.display import clear_output
from time import sleep
from google.colab.patches import cv2_imshow

# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', _verbose=False)
fn="/content/DataSet/RoadTrafic.mp4"
cap = cv2.VideoCapture(fn)
if (cap.isOpened()== False): 
  print("Error opening video stream or file")

# Membaca video sampai selesai
n=0
while(cap.isOpened()):
  #Dalam contoh ini video di baca sampai dengan frame ke 20
  n=n+1 
  if n>20:
    break
  # Capture frame-by-frame
  ret, frame = cap.read()
  if ret == True:
    results = model(frame)
    #Render Boundingbox dan kelas ke frame
    results.render()
    print("frame : ",n)
    cv2_imshow(results.imgs[0])
    sleep(1)
    clear_output(wait=True)
````

### 5.4.2. Deteksi Kendaraan Di Jalan Tol Versi Komputer Dekstop
Di bawah ini adalah contoh program yang terapkan pada Komputer lokal.

````python
#Listing 5.7. 
import cv2 
import torch
import numpy as np 


# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', _verbose=False)
fn="..\\..\\DataSetPelatihanV5\\RoadTrafic.mp4"
cap = cv2.VideoCapture(fn)
if (cap.isOpened()== False): 
  print("Error opening video stream or file")

# Baca sam pai selesai
while(cap.isOpened()):
  # Capture frame-by-frame
  ret, frame = cap.read()
  if ret == True:
    results = model(frame)
    results.render()
    cv2.imshow('Frame',results.imgs[0])
    # Tekan Tombol Q Untuk Keluar
    if cv2.waitKey(25) & 0xFF == ord('q'):
      break

  # Keluar dari loop
  else: 
    break
cap.release()
# Hapus Semua frame
cv2.destroyAllWindows()
````
