```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
```

```python
# Background removal with opencv
fb = cv2.imread('img/lenna.png')

fb = cv2.resize(fb, [640,480], interpolation = cv2.INTER_AREA)
fb=np.double(fb)/255
cap = cv2.VideoCapture(0)
f=0
gray2=[]
while True:
    ret, f = cap.read()
    [b,k,d]=fb.shape
    print(b,k)
    
    
    
    
    hsv= np.double(cv2.cvtColor(f, cv2.COLOR_BGR2HSV))
    h=hsv[:,:,0]       
    h = h*2
    h[h<60]=0
    h[h>180]=0
    h[h>0]=1
    h=1-h
    
    fw =np.double(f)
    fw =fw/255
    fw[:,:,0]=fw[:,:,0]*h+fb[:,:,0]*(1-h)
    fw[:,:,1]=fw[:,:,1]*h+fb[:,:,1]*(1-h)
    fw[:,:,2]=fw[:,:,2]*h+fb[:,:,2]*(1-h)

    cv2.imshow('frame2', fw)
    
     

     
    if cv2.waitKey(1) == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
